```python
import numpy as np
import pandas as pd
df = pd.read_csv("titles.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df1 = pd.read_csv("cast.csv")
df1.head()
```

```python
df2 = pd.read_csv("release_dates.csv")
df2.head()
```

```python
df3 = df2[df2['title'].str.contains('Christmas', case=False, na=False)]
df4 = df3[df3['country'] == 'USA']
df4['release_date'] = pd.to_datetime(df4['date'])
df4['release_month'] =df4['release_date'].dt.month
df5 = df4['release_month'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
df5.plot(kind='bar', color='blue', edgecolor='black')
plt.title('Number of Christmas Movies Released by Month (USA)')
plt.xlabel('Month')
plt.ylabel('Number of Movies Released')
plt.xticks(rotation=0)
plt.show()
```

```python
df3 = df2[df2['title'].str.contains('The Hobbit', na=False, case=False)]
df4 = df3[df3['country'] == 'USA']
df4['release_date'] = pd.to_datetime(df4['date'])
df4['release_month'] = df4['release_date'].dt.month
df5 = df4['release_month'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
df5.plot(kind='bar', color='blue', edgecolor='black')
plt.title('Number of "The Hobbit" Movies Released by Month (USA)')
plt.xlabel('Month')
plt.ylabel('Number of Movies Released')
plt.xticks(rotation=0)
plt.show()
```

```python
df3 = df2[df2['title'].str.contains('Romance', case=False, na=False)]
df4 = df3[df3['country'] == 'USA']
df4['release_date'] = pd.to_datetime(df4['date'])
df4['release_day_of_week'] = df4['release_date'].dt.day_name()
df5 = df4['release_day_of_week'].value_counts().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.figure(figsize=(10, 6))
df5.plot(kind='bar', color='blue', edgecolor='black')
plt.title('Day of the Week for "Romance" Movie Releases (USA)')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Movies Released')
plt.xticks(rotation=45)
plt.show()
```
